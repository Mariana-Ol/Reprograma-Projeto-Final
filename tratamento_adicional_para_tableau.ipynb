```python
import pandas as pd
```

```python
colunas = [ 'ID_UF', 'ID_AREA', 'IN_PUBLICA', 'ID_LOCALIZACAO', 'PROFICIENCIA_MT_SAEB']
```

```python
df = pd.read_csv('TS_ALUNO_9EF.csv', sep=';', usecols = colunas)
```

```python
df
```

```python
df.info()
```

```python
df.columns = ['UF', 'Area', 'Pública ou privada?', 'Urbana ou rural?', 'Proficiência em matemática']
```

```python
df
```

```python
df['UF'].unique()
```

```python
df['UF'] = df['UF'].replace({11 : 'RO', 12 : 'AC', 13 : 'AM', 14 : 'RR', 15 : 'PA', 16 : 'AP', 17 : 'TO',  21: 'MA', 22 : 'PI', 23 : 'CE', 24 : 'RN', 25 : 'PB', 26 : 'PE',
            27 : 'AL', 28 : 'SE', 29 : 'BA', 31 : 'MG', 32 : 'ES', 33 : 'RJ', 35 : 'SP', 41 : 'PR', 42 : 'SC', 43 : 'RS',
            50 : 'MS', 51 : 'MT', 52: 'GO', 53 : 'DF'})
```

```python
df['Area'] = df['Area'].replace({1 : 'Capital', 2 : 'Interior'})
```

```python
df['Pública ou privada?'] = df['Pública ou privada?'].replace({0 : 'Privada', 1 : 'Pública'})
```

```python
df['Urbana ou rural?'] = df['Urbana ou rural?'].replace({1 : 'Urbana', 2 : 'Rural'})
```

```python
df
```

```python

df = df.drop_duplicates()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['Proficiência em matemática'])
```

```python
df = df[df['Pública ou privada?'].str.match('Pública')]
df
```

```python
df.to_csv("df_SAEB_com_todas_as_regioes.csv")
```
